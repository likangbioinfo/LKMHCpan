```python
import matplotlib.pyplot as plt
def plot_bar(dict1,dict2,Soft,outplot):
    # 获取所有键（x 轴标签）
    keys = list(dict1.keys())
    # 获取每组字典的值
    values1 = list(dict1.values())
    values2 = list(dict2.values())
    plt.figure(figsize=(3, 3))
    # 设置条形图的宽度
    bar_width = 0.36

    # 设置 x 轴的位置
    index = range(len(keys))

    # 创建条形图
    plt.bar(index, values1, width=bar_width, label='NetMHCpan4.1_EL+RNA', alpha=0.8,color='#008000',edgecolor='black',linewidth=0.5)
    plt.bar([i + bar_width for i in index], values2, width=bar_width, label='LKMHCpan+RNA', alpha=0.8,color='orange',edgecolor='black',linewidth=0.5)

    # 添加 x 轴标签
    #plt.xlabel('Sample')
    plt.ylabel('Number of identified patient')
    plt.title(Soft)
    # 设置 x 轴刻度位置和标签
    plt.xticks([i + bar_width / 2 for i in index], keys)
    # 添加图例
    plt.legend()
    # 显示图形
    plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
    plt.savefig(outplot)
    plt.show()
```

```python
##将同类型画到一起
dict_LK={"Top30":0.75,"Top50":1}
dict_Net={"Top30":0.625,"Top50":0.875}
print(dict_LK,dict_Net)

plot_bar(dict_Net,dict_LK,"patient number","/life/Project/liqing/02.Work/01.LNEA.toolkit/02.Neo_immunity/01.TESLA/00.data_example/LKMHCpan2_TESLA_Data_CD8_Top_patient.svg")
```
